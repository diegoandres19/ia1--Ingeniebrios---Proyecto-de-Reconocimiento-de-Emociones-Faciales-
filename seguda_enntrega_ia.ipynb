```python

!pip install --upgrade pip
!pip install kagglehub tensorflow scikit-learn matplotlib pandas joblib seaborn --quiet


import kagglehub
from pathlib import Path
import zipfile, shutil, os, sys, random
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
print("Librerías listas.")

DATASET_ID = "sujaykapadnis/emotion-recognition-dataset"
print("Descargando dataset:", DATASET_ID)
download_path = kagglehub.dataset_download(DATASET_ID)  # devuelve ruta
print("kagglehub returned:", download_path)

p = Path(download_path)
extract_dir = Path("/content/dataset")
extract_dir.mkdir(parents=True, exist_ok=True)


zip_files = list(p.glob("*.zip"))
if zip_files:
    print("Encontrado zip:", zip_files[0])
    with zipfile.ZipFile(str(zip_files[0]), "r") as zf:
        zf.extractall(extract_dir)
else:
    try:
        for child in p.iterdir():
            target = extract_dir / child.name
            if child.is_dir():
                if target.exists(): shutil.rmtree(target)
                shutil.copytree(child, target)
            else:
                shutil.copy2(child, target)
    except Exception as e:
        print("Advertencia al copiar archivos automáticamente:", e)

print("Contenido de /content/dataset (top):", [x.name for x in extract_dir.iterdir()][:20])


dataset_root = extract_dir
csv_candidates = list(dataset_root.glob("**/*.csv"))
if not csv_candidates:
    raise FileNotFoundError("No se encontró CSV dentro de /content/dataset")
csv_path = csv_candidates[0]
print("Usando CSV:", csv_path)

df = pd.read_csv(csv_path)
print("CSV shape:", df.shape)
print("Columns:", df.columns.tolist())

# Normalizar nombres de columnas a 'path' y 'label'
if 'path' not in df.columns:
    for c in df.columns:
        if 'path' in c.lower() or 'image' in c.lower() or 'file' in c.lower():
            df = df.rename(columns={c: 'path'})
            break
if 'label' not in df.columns:
    for c in df.columns:
        if 'label' in c.lower() or 'emotion' in c.lower() or 'class' in c.lower():
            df = df.rename(columns={c: 'label'})
            break



!pip install --upgrade pip
!pip install kagglehub tensorflow scikit-learn matplotlib pandas joblib seaborn --quiet


import kagglehub
from pathlib import Path
import zipfile, shutil, os, sys, random
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
print("Librerías listas.")


# Parámetros

DATASET_ID = "sujaykapadnis/emotion-recognition-dataset"
EXTRACT_DIR = Path("/content/dataset")
TOTAL_TARGET = 700   # total imágenes a usar
TEST_RATIO = 0.20   # test size (en la práctica con 10 -> 2 test, 8 train)
RANDOM_STATE = 42
IMG_SIZE = (224, 224)
BATCH_SIZE = 8


print("Descargando dataset:", DATASET_ID)
download_path = kagglehub.dataset_download(DATASET_ID)
print("kagglehub returned:", download_path)

p = Path(download_path)
EXTRACT_DIR.mkdir(parents=True, exist_ok=True)


zip_files = list(p.glob("*.zip"))
if zip_files:
    print("Encontrado zip:", zip_files[0])
    with zipfile.ZipFile(str(zip_files[0]), "r") as zf:
        zf.extractall(EXTRACT_DIR)
else:
    try:
        for child in p.iterdir():
            target = EXTRACT_DIR / child.name
            if child.is_dir():
                if target.exists(): shutil.rmtree(target)
                shutil.copytree(child, target)
            else:
                shutil.copy2(child, target)
    except Exception as e:
        print("Advertencia al copiar archivos automáticamente:", e)

print("Contenido top de /content/dataset:", [x.name for x in EXTRACT_DIR.iterdir()][:30])


dataset_root = EXTRACT_DIR
csv_candidates = list(dataset_root.glob("**/*.csv"))
if not csv_candidates:
    raise FileNotFoundError("No se encontró CSV dentro de /content/dataset")
csv_path = csv_candidates[0]
print("Usando CSV:", csv_path)

df = pd.read_csv(csv_path)
print("CSV shape:", df.shape)
print("Columns:", df.columns.tolist())

# Normalizar nombres de columnas a 'path' y 'label'
if 'path' not in df.columns:
    for c in df.columns:
        if 'path' in c.lower() or 'image' in c.lower() or 'file' in c.lower():
            df = df.rename(columns={c: 'path'})
            break
if 'label' not in df.columns:
    for c in df.columns:
        if 'label' in c.lower() or 'emotion' in c.lower() or 'class' in c.lower():
            df = df.rename(columns={c: 'label'})
            break

df.columns = [c.strip() for c in df.columns]
assert 'path' in df.columns and 'label' in df.columns, f"CSV debe tener columnas 'path' y 'label'. Columns: {df.columns}"


# Encontrar images_root y construir full_path

unique_labels = sorted(df['label'].unique())
print("Etiquetas detectadas en CSV:", unique_labels)

def find_images_root(base_dir, labels):
    candidates = [base_dir] + [x for x in base_dir.iterdir() if x.is_dir()]
    for candidate in candidates:
        try:
            subdirs = [d.name.lower() for d in candidate.iterdir() if d.is_dir()]
        except Exception:
            subdirs = []
        match_count = sum(1 for lbl in labels if lbl.lower() in subdirs)
        if match_count >= max(1, len(labels)//3):
            return candidate
    return base_dir

images_root = find_images_root(dataset_root, unique_labels)
print("Usando images_root =", images_root)

df['full_path'] = df['path'].apply(lambda p: str(images_root / p))

from pathlib import Path
exists_mask = df['full_path'].apply(lambda x: Path(x).exists())
missing_count = (~exists_mask).sum()
print("Archivos faltantes:", missing_count, "de", len(df))
if missing_count:
    print("Ejemplos faltantes (hasta 5):", df.loc[~exists_mask, 'full_path'].head(5).tolist())

df = df[exists_mask].reset_index(drop=True)
print("Usamos", len(df), "archivos encontrados para procesar.")


# Muestreo: mínimo 1 por clase + completar hasta TOTAL_TARGET

TOTAL = TOTAL_TARGET
classes_all = sorted(df['label'].unique())
n_classes = len(classes_all)
print("Clases:", classes_all, "n_classes =", n_classes)

if len(df) <= TOTAL:
    sampled_df = df.copy()
    print(f"Dataset tiene <= {TOTAL} imágenes. Se usan todas ({len(df)}).")
else:
    # 1 por clase si posible
    per_class = []
    for c in classes_all:
        c_df = df[df['label'] == c]
        if len(c_df) >= 1:
            per_class.append(c_df.sample(n=1, random_state=RANDOM_STATE))
        else:
            print(f"Advertencia: clase {c} no tiene ejemplos.")
    sampled_df = pd.concat(per_class).reset_index(drop=True)
    remaining = TOTAL - len(sampled_df)
    if remaining > 0:
        remaining_pool = df.drop(sampled_df.index, errors='ignore')
        if len(remaining_pool) <= remaining:
            extra = remaining_pool.sample(n=len(remaining_pool), random_state=RANDOM_STATE)
        else:
            extra = remaining_pool.sample(n=remaining, random_state=RANDOM_STATE)
        sampled_df = pd.concat([sampled_df, extra]).reset_index(drop=True)

# mezclar y reasignar
sampled_df = sampled_df.sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)
print("Muestreo final. Total muestras:", len(sampled_df))
print(sampled_df['label'].value_counts())

df = sampled_df.copy()


# Re-mapeo de clases y conteo

classes = sorted(df['label'].unique())
class_to_idx = {c: i for i, c in enumerate(classes)}
idx_to_class = {i: c for c, i in class_to_idx.items()}
print("Clases usadas (orden):", classes)


# Train/Test split (sin validation) - estratificado

from sklearn.model_selection import train_test_split
try:
    train_df, test_df = train_test_split(df, test_size=TEST_RATIO, stratify=df['label'], random_state=RANDOM_STATE)
    print("Split estratificado OK.")
except Exception as e:
    print("Advertencia: split estratificado falló (pocos datos). Mensaje:", e)
    train_df, test_df = train_test_split(df, test_size=TEST_RATIO, random_state=RANDOM_STATE, shuffle=True)
    print("Split aleatorio realizado como fallback.")

print("Tamaños -> train:", len(train_df), "test:", len(test_df))
print("Train label counts:\n", train_df['label'].value_counts())


# Crear tf.data datasets
import tensorflow as tf
AUTOTUNE = tf.data.AUTOTUNE

def df_to_tf_dataset(df_in, shuffle=True):
    paths = df_in['full_path'].values
    labels = df_in['label'].map(class_to_idx).values
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    def _load(path, label):
        path = tf.cast(path, tf.string)
        img = tf.io.read_file(path)
        img = tf.image.decode_image(img, channels=3, expand_animations=False)
        img = tf.image.resize(img, IMG_SIZE)
        img = tf.keras.applications.mobilenet_v2.preprocess_input(img)
        label = tf.one_hot(label, depth=len(classes))
        return img, label
    ds = ds.map(lambda p, l: tf.py_function(func=_load, inp=[p, l], Tout=(tf.float32, tf.float32)), num_parallel_calls=AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(buffer_size=2048, seed=RANDOM_STATE)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

train_ds = df_to_tf_dataset(train_df, shuffle=True)
test_ds = df_to_tf_dataset(test_df, shuffle=False)
print("Datasets TF listos. batches:", "train", len(list(train_ds)), "test", len(list(test_ds)))


# Extraccion embeddings con MobileNetV2

from tensorflow.keras import layers, models

base = tf.keras.applications.MobileNetV2(input_shape=IMG_SIZE+(3,), include_top=False, weights='imagenet')
base.trainable = False
inp = layers.Input(shape=IMG_SIZE+(3,))
x = base(inp, training=False)
x = layers.GlobalAveragePooling2D()(x)
extractor = models.Model(inp, x)
print("Extractor listo. Output shape:", extractor.output_shape)

def dataset_to_features(ds):
    Xs, ys = [], []
    for batch_imgs, batch_labels in ds:
        feats = extractor.predict(batch_imgs, verbose=0)
        Xs.append(feats)
        ys.append(np.argmax(batch_labels.numpy(), axis=1))
    if len(Xs) == 0:
        return np.empty((0, extractor.output_shape[1])), np.empty((0,), dtype=int)
    X = np.vstack(Xs)
    y = np.concatenate(ys)
    return X, y

print("Extrayendo features (rápido por pocas muestras)...")
X_train, y_train = dataset_to_features(train_ds)
X_test, y_test = dataset_to_features(test_ds)
print("Features shapes:", X_train.shape, X_test.shape)


# Escalado con StandardScaler

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
if X_train.shape[0] > 0:
    X_train_s = scaler.fit_transform(X_train)
    X_test_s = scaler.transform(X_test) if X_test.shape[0] > 0 else np.empty_like(X_test)
else:
    X_train_s = X_train; X_test_s = X_test
joblib.dump(scaler, '/content/feature_scaler.pkl')
print("Scaler guardado en /content/feature_scaler.pkl")


# Entrenar clasificadores sklearn

from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

models_dict = {
    'GaussianNB': GaussianNB(),
    'DecisionTree': DecisionTreeClassifier(random_state=RANDOM_STATE),
    'RandomForest': RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1),
    'SVM': SVC(kernel='rbf', probability=True, random_state=RANDOM_STATE)
}

results = {}
for name, clf in models_dict.items():
    print(f"\n--- Training {name} ---")
    try:
        clf.fit(X_train_s, y_train)
        if X_test_s.shape[0] > 0:
            y_pred = clf.predict(X_test_s)
            acc = accuracy_score(y_test, y_pred)
            # obtener etiquetas presentes en y_test para safe reporting
            unique_test = sorted(np.unique(y_test))
            target_names_test = [classes[i] for i in unique_test]
            print(f"{name} test accuracy: {acc:.4f}")
            print(classification_report(y_test, y_pred, labels=unique_test, target_names=target_names_test, zero_division=0))
            # guardar y_pred en results para plotting
            results[name] = {'clf': clf, 'acc': acc, 'y_pred': y_pred}
        else:
            print(f"{name}: no hay muestras de test para evaluar.")
            results[name] = {'clf': clf, 'acc': float('nan'), 'y_pred': []}
        joblib.dump(clf, f"/content/{name}_mobilenet_features.joblib")
        print(f"Guardado: /content/{name}_mobilenet_features.joblib")
    except Exception as e:
        print(f"Error entrenando/evaluando {name}: {e}")


# MLP en Keras (sobre embeddings)
from tensorflow.keras import layers, models, callbacks, optimizers

num_classes = len(classes)
def make_mlp_model(input_dim):
    i = layers.Input(shape=(input_dim,))
    x = layers.Dropout(0.2)(i)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.2)(x)
    o = layers.Dense(num_classes, activation='softmax')(x)
    m = models.Model(i, o)
    m.compile(optimizer=optimizers.Adam(1e-3), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return m

if X_train_s.shape[0] > 0:
    mlp = make_mlp_model(X_train_s.shape[1])
    mlp.summary()
    cb_list = [
        callbacks.EarlyStopping(monitor='loss', patience=10, restore_best_weights=True),
        callbacks.ReduceLROnPlateau(monitor='loss', factor=0.5, patience=4, min_lr=1e-6),
        callbacks.ModelCheckpoint('/content/mlp_best.h5', monitor='loss', save_best_only=True)
    ]

    history = mlp.fit(X_train_s, y_train, validation_data=None, epochs=50, batch_size=4, callbacks=cb_list, verbose=1)
    mlp.save('/content/mlp_mobilenet_features.h5')
    print("MLP guardado en /content/mlp_mobilenet_features.h5")
else:
    print("No hay datos para entrenar MLP.")


# Matrices de confusión y reportes

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def plot_cm(y_true, y_pred, title):
    unique = sorted(np.unique(y_true))
    if len(unique) == 0 or len(y_pred) == 0:
        print("No hay suficientes datos para calcular matriz de confusión para", title)
        return
    labels_names = [classes[i] for i in unique]
    cm = confusion_matrix(y_true, y_pred, labels=unique)
    cmn = cm.astype('float') / (cm.sum(axis=1)[:, np.newaxis] + 1e-9)
    plt.figure(figsize=(6,5))
    sns.heatmap(cmn, annot=True, fmt='.2f', xticklabels=labels_names, yticklabels=labels_names, cmap='Blues')
    plt.title(title)
    plt.ylabel('True')
    plt.xlabel('Predicted')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.show()

# Plot for sklearn models
for name, info in results.items():
    y_pred = info.get('y_pred', [])
    if len(y_pred) > 0:
        plot_cm(y_test, y_pred, f"{name} (acc={info['acc']:.3f})")
    else:
        print(f"No hay predicciones para {name} para mostrar matriz.")

# MLP predictions
if 'mlp' in locals() and X_test_s.shape[0] > 0:
    y_pred_mlp_probs = mlp.predict(X_test_s)
    y_pred_mlp = np.argmax(y_pred_mlp_probs, axis=1)
    acc_mlp = accuracy_score(y_test, y_pred_mlp)
    unique_test = sorted(np.unique(y_test))
    target_names_test = [classes[i] for i in unique_test]
    print("MLP test accuracy:", acc_mlp)
    print(classification_report(y_test, y_pred_mlp, labels=unique_test, target_names=target_names_test, zero_division=0))
    plot_cm(y_test, y_pred_mlp, f"MLP Keras (acc={acc_mlp:.3f})")
else:
    print("No se generaron predicciones MLP (pocos datos o MLP no entrenado).")


all_results = {k: v.get('acc', float('nan')) for k, v in results.items()}
if 'mlp' in locals() and X_test_s.shape[0] > 0:
    all_results['MLP_Keras'] = accuracy_score(y_test, np.argmax(mlp.predict(X_test_s), axis=1))
print("Resumen accuracies:", all_results)
with open('/content/accuracies.json', 'w') as f:
    import json
    json.dump(all_results, f)
print("Accuracies guardadas en /content/accuracies.json")

print("Script completado. Modelos y artefactos guardados en /content/")

```

# Ejemplo tomando 5 imagenes aleatorias y siendo clasificadas por cada modelo

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import joblib
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing import image
import os

# === CONFIGURACIÓN ===
data_dir = "/root/.cache/kagglehub/datasets/sujaykapadnis/emotion-recognition-dataset/versions/1/dataset"
model_paths = {
    'GaussianNB': '/content/GaussianNB_mobilenet_features.joblib',
    'DecisionTree': '/content/DecisionTree_mobilenet_features.joblib',
    'RandomForest': '/content/RandomForest_mobilenet_features.joblib',
    'SVM': '/content/SVM_mobilenet_features.joblib',
    'MLP_Keras': '/content/mlp_mobilenet_features.h5'
}

# === CARGAR SCALER ===
scaler = joblib.load("/content/feature_scaler.pkl")

# === PREPARAR EXTRACTOR DE FEATURES ===
mobilenet = MobileNetV2(weights="imagenet", include_top=False, pooling="avg")

# === CARGAR MODELOS ===
sklearn_models = {}
for name, path in model_paths.items():
    if name == 'MLP_Keras':
        keras_model = tf.keras.models.load_model(path)
    else:
        sklearn_models[name] = joblib.load(path)

# === FUNCIONES ===
def load_and_preprocess(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    return img_array

def extract_features(img_array):
    features = mobilenet.predict(img_array)
    features = scaler.transform(features)
    return features

# === TOMAR 5 IMÁGENES ALEATORIAS ===
labels = sorted(os.listdir(data_dir))  # asumimos carpetas = clases
all_images = []
for label in labels:
    folder = os.path.join(data_dir, label)
    if os.path.isdir(folder):
        imgs = [os.path.join(folder, i) for i in os.listdir(folder)]
        all_images.extend(imgs)

sample_images = random.sample(all_images, 5)

# === CLASIFICAR Y MOSTRAR ===
plt.figure(figsize=(20, 8))
for i, img_path in enumerate(sample_images):
    img_array = load_and_preprocess(img_path)
    features = extract_features(img_array)

    plt.subplot(1, 5, i + 1)
    img = image.load_img(img_path)
    plt.imshow(img)
    plt.axis("off")

    preds_text = ""
    # sklearn models
    for name, model in sklearn_models.items():
        pred_idx = model.predict(features)[0]
        pred_label = labels[pred_idx]
        preds_text += f"{name}: {pred_label}\n"
    # MLP Keras
    pred_idx = np.argmax(keras_model.predict(features), axis=1)[0]
    pred_label = labels[pred_idx]
    preds_text += f"MLP_Keras: {pred_label}"

    plt.title(preds_text, fontsize=9)

plt.tight_layout()
plt.show()
```
